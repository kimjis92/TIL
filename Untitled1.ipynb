```python
import cv2
import numpy as np
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator

# 얼굴과 눈동자 검출기 생성
face_cascade = cv2.CascadeClassifier('haarcascade_frontface.xml')
clf = load_model("knaivesji.h5") #모델 불러오기
def findFaces(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray)
    face_coords = []
    for (x,y,w,h) in faces:
        face_coords.append((x, y, w, h))
    return face_coords

cap = cv2.VideoCapture("knivesout.mp4")

data_gen=ImageDataGenerator(rescale=1./ 255)

ch_count=0
cr_count=0
ev_count=0
mi_count=0
st_cout=0
to_count=0
un_count=0


while True:
    ret, frame = cap.read()
    if not ret : break
    img1 = frame.copy()
    # 얼굴 검출해서 오목/볼록 렌즈 효과로 왜곡 적용
    faces = findFaces(frame)
    for face in faces:
        x,y,w,h = face
        roi = img1[y:y+h, x:x+w]
        roi2=cv2.resize(roi, (150, 150), interpolation=cv2.INTER_AREA)
        roi3= roi2.reshape(1, 150, 150, 3)
        pre=clf.predict(roi3)
       # print(pre)
        if np.argmax(pre)==0:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame,'chirstopher',(x,y-30),cv2.FONT_ITALIC, 1., (255,255,255))
            ch_count+=1
            if ch_count%10==0:
                cv2.imwrite('./immo/ch'+str(ch_count)+'.jpg',frame)
            
        elif np.argmax(pre)==1:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame,'creig',(x,y-30),cv2.FONT_ITALIC, 1., (255,255,255))
            cr_count+=1
            if cr_count%10==0:
                cv2.imwrite('./immo/cr'+str(cr_count)+'.jpg',frame)
                
        elif np.argmax(pre)==2:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame,'evans',(x,y-30),cv2.FONT_ITALIC, 1., (255,255,255))
            ev_count+=1
            if ev_count%10==0:
                cv2.imwrite('./immo/ev'+str(ev_count)+'.jpg',frame)
                        
        elif np.argmax(pre)==3:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame,'michael',(x,y-30),cv2.FONT_ITALIC, 1., (255,255,255))
            mi_count+=1
            if mi_count%10==0:
                cv2.imwrite('./immo/mi'+str(ch_count)+'.jpg',frame)
        
        elif np.argmax(pre)==4:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame,'stanfield',(x,y-30),cv2.FONT_ITALIC, 1., (255,255,255))
            st_count+=1
            if st_count%10==0:
                cv2.imwrite('./immo/st'+str(ch_count)+'.jpg',frame)
                    
        elif np.argmax(pre)==5:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame,'toni',(x,y-30),cv2.FONT_ITALIC, 1., (255,255,255))
            to_count+=1
            if to_count%10==0:
                cv2.imwrite('./immo/to'+str(ch_count)+'.jpg',frame)
                        
        elif np.argmax(pre)==6:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(frame,'unknown',(x,y-30),cv2.FONT_ITALIC, 1., (255,255,255))
            un_count+=1
            if un_count%10==0:
                cv2.imwrite('./immo/un'+str(ch_count)+'.jpg',frame)
            
    #cv2.imwrite('./immo', frame)        
    cv2.imshow('w', frame)
    if cv2.waitKey(1) == 13: break
        
cap.release()
cv2.destroyAllWindows()
```
